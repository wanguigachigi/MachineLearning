```python
pred_data=data[['YEAR','BOROUGH','CONTRIBUTING FACTOR VEHICLE 1','VEHICLE TYPE CODE 1','NUMBER OF PERSONS INJURED','NUMBER OF PERSONS KILLED']].sort_values(by='YEAR',ascending=True)
```

```python
pred_data['OUTCOME']= pred_data['NUMBER OF PERSONS KILLED'].apply(lambda x:1 if x == True else 0)
```

```python
pred_data.groupby(['BOROUGH','YEAR','OUTCOME']).sum()
```

```python
pred_data.drop(['BOROUGH','CONTRIBUTING FACTOR VEHICLE 1','VEHICLE TYPE CODE 1'], axis=1, inplace=True)
```

```python
pred_data.dropna(axis=0,inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score, classification_report
```

```python
pred_data['OUTCOME'].value_counts()
```

```python
X = pred_data.drop(columns='OUTCOME',axis=1)
Y = pred_data['OUTCOME']
```

```python
print(X)
```

```python
print(Y)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standard_data = scaler.transform(X)
```

```python
X= standard_data
```

```python
print(X)
print(Y)
```

```python
X_train, X_test,Y_train,Y_test =train_test_split(X,Y,test_size=0.5,stratify=Y,random_state=101)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train,Y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
print(classification_report(Y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(Y_test, predictions))
```
