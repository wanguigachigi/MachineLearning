```python
X = data[['SHAPE_Length','SHAPE_Area']]
X = X[:int(len(data)-1)]
```

```python
y = data['Obesity']
y = y[:int(len(data)-1)]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
prediction = lr.predict(X_test)
```

```python
prediction
```

```python
error = y_test - prediction
```

```python
error
```

```python
pred_data = pd.DataFrame({'prediction':prediction,'error_margin':error})
```

```python
pred_data
```

```python
new_data = pd.concat([X_test,y_test,pred_data],axis=1)
```

```python
new_data
```

```python
print('The score is: ',lr.score(X,y))
```
