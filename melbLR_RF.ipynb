```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('melb_data.csv')
```

```python
data.fillna(0,inplace=True)
```

```python
data['Propertycount'].value_counts()
```

```python
quantiles = data['Price'].quantile([0.25, 0.5, 0.75])
```

```python
quantiles
```

```python
def cat_by_price(row):
    if row['Price'] <= quantiles.iloc[0]:
        return 'low'
    elif row['Price'] <= quantiles.iloc[1]:
        return 'mid-range'
    elif row['Price'] <= quantiles.iloc[2]:
        return 'high'
    else:
        return 'premium'
```

```python
data['Price_category'] = data.apply(cat_by_price, axis=1)
```

```python
data[['Price', 'Price_category']]
```

```python
data['Price_category'].value_counts()
```

```python
 low_priced_homes = data.query('Price_category == "low"')
```

```python
 mid_priced_homes = data.query('Price_category == "mid-range"')
```

```python
 high_priced_homes = data.query('Price_category == "high"')
```

```python
 premium_priced_homes = data.query('Price_category == "premium"')
```

```python
data
```

```python
data['YearBuilt'] = pd.to_numeric(pd.to_datetime(data['YearBuilt'], unit='s'))
```

```python
data['YearBuilt']
```

```python
# plotting each price range in the same figure
sns.ecdfplot(data=low_priced_homes, x='YearBuilt', color='blue', label='Low')
sns.ecdfplot(data=mid_priced_homes, x='YearBuilt', color='yellow', label='Mid-Range')
sns.ecdfplot(data=high_priced_homes, x='YearBuilt', color='green', label='High')
sns.ecdfplot(data=premium_priced_homes, x='YearBuilt', color='red', label='Premium')

plt.title('Plot for the various price points')
plt.ylabel('ECDF')
plt.xlabel('Year Built')
plt.grid(True)
plt.legend(title='Price Points')
```

```python
df = data.drop(['Suburb','Address','Method','SellerG','Date','Postcode','CouncilArea','Regionname','Price_category'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
```

```python
df['Type'] = enc.fit_transform(df['Type'])
```

```python
df['Type'].value_counts()
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
```

```python
X = df.drop('Price', axis=1)
```

```python
y = df['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
pipe_lr = Pipeline([('scaler', StandardScaler()),('model',LinearRegression())])
```

```python
pipe_lr.fit(X_train,y_train)
```

```python
model_lr = pipe_lr.predict(X_test)
```

```python
model_lr
```

```python
pipe_lr.score(X_test,y_test)
```

```python
# The above linear model performed so bad, lets check other models
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

```python
pipe_rf  = Pipeline([('scaler', StandardScaler()),('model',RandomForestRegressor())])
```

```python
pipe_rf.fit(X_train,y_train)
```

```python
model_rf = pipe_rf.predict(X_test)
```

```python
pipe_rf.score(X_test,y_test)
```

```python
X_test = pd.DataFrame(X_test,columns=['Rooms','Type','Distance' ,'Bedroom2','Bathroom','Car','Landsize','BuildingArea','YearBuilt','Lattitude','Longtitude','Propertycount'])
```

```python
y_test = pd.DataFrame(y_test,columns=['Price'])
```

```python
model_lr = pd.DataFrame(model_lr, columns=['Predicted_lr'])
```

```python
model_rf = pd.DataFrame(model_rf,columns=['Predicted_rf'])
```

```python
new_data = pd.concat([X_test,y_test,model_lr,model_rf],axis=1)
```

```python
new_data
```

```python
sns.ecdfplot(data=new_data, x='Price', color='red', label='Actual Price')
sns.ecdfplot(data=new_data, x='Predicted_lr', color='yellow', label='Linear Regression')
sns.ecdfplot(data=new_data, x='Predicted_rf', color='green', label='Random Forest')

plt.title('Actual Price and Predicted prices')
plt.ylabel('ECDF')
plt.xlabel('Predicted Prices')
plt.grid(True)
plt.legend(title='Different Models')
```
